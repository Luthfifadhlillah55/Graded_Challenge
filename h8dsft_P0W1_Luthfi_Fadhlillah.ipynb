## Deskripsi Dataset
This data set includes 721 Pokemon, including their number, name, first and second type, and basic stats: HP, Attack, Defense, Special Attack, Special Defense, and Speed.

The feature of this data is:

- #: ID for each pokemon
- Name: Name of each pokemon
- Type 1: Each pokemon has a type, this determines weakness/resistance to attacks
- Type 2: Some pokemon are dual type and have 2
- Total: sum of all stats that come after this, a general guide to how strong a pokemon is
- HP: hit points, or health, defines how much damage a pokemon can withstand before fainting
- Attack: the base modifier for normal attacks (eg. Scratch, Punch)
- Defense: the base damage resistance against normal attacks
- SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
- SP Def: the base damage resistance against special attacks
- Speed: determines which pokemon attacks first each round

## Data Cleaning & Missing Value

```python
data.head(10)
```

Dari data diatas, terdapat penulisan nama pokemon yang tidak perlu ditulis, seperti "CharizardMega Charizard X" dan "CharizardMega Charizard Y" yang seharusnya bisa ditulis "Mega Charizard X/Y" saja, karena charizard adalah nama 1 pokemon, dan pengulangan kata charizard rasanya tidak perlu dilakukan dalam pemberian sebuah nama. Jadi saya akan menghapus kata "charizard" sebelum kata "Mega"

```python
data.Name = data.Name.str.replace(".*(?=Mega)", "")
data.head(10)
```

Jika kita melihat baris 5, ada tipe NaN di column Type 2. Kita dapat memilih untuk menghapus atau mengisi data. Namun dalam hal ini jika kita menghapus nilai NaN, maka akan mengacaukan data. Lebih baik mengisi NaN dengan "Fire" karena Charmander merupakan pokemon type Fire sebelum berevolusi menjadi charmeleon dan charizard. Begitupun untuk nilai NaN pada Type 2 yang lainnya akan diisi dengan kata yang berada di Type 1.

```python
data['Type 2'].fillna(data['Type 1'], inplace=True)
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['Type 2'].isnull().any()
```

Pada awalnya Column "Type 2" memiliki nilai NaN, namun setelah data NaN yang ada diisi berdasarkan data yang ada di "Type 1", maka sekarang dataset tidak memiliki nilai NaN lagi.

## Eksplorasi Data
Melihat semua unique types di Type 1 dan Type 2.

## Pokemon Terkuat
Pokemon manakah yang merupakan Pokemon terkuat berdasarkan dataset yang ada?

```python
strongest = data.sort_values(by='Total', ascending=False) #sorting baris dalam urutan menurun
strongest.drop_duplicates(subset=['Type 1'],keep='first')
```

Berdasarkan data diatas kita dapat mengetahui pokemon mana yang terkuat. Mega Rayquaza adalah pokemon terkuat dan bertipe Dragon. Dan juga kita tahu bahwa 10/18 Pokemon Terkuat berdasarkan jenisnya adalah Legendary Pokemon.

## Pokemon Terlemah
Pokemon manakah yang merupakan pokemon terlemah berdasarkan dataset yang ada?

```python
weakest = data.sort_values(by='Total') #sorting baris dalam urutan menurun
weakest.drop_duplicates(subset=['Type 1'],keep='first')
```

Berdasarkan data diatas, pokemon terlemahnya adalah Sunkern yang merupakan tipe Grass, dan pokemon terlemah bukan merupakan pokemon legendaris.
